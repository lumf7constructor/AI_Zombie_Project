# Phase 2: AI Zombie Game - Algorithm Comparison & Metrics

This notebook compares different minimax-based algorithms used in the AI Zombie game scenarios. We analyze performance metrics, efficiency, and outcomes across 4 different strategy implementations.

## Scenarios Compared:
1. **Both MINIMAX (d=60)** - Both players use deep lookahead
2. **Ambush Monster** - Monster uses MINIMAX, Prey uses A*
3. **Both MINIMAX (d=3)** - Both players use shallow lookahead
4. **Evasive Prey** - Prey uses MINIMAX, Monster uses A*

```python
# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pathlib import Path

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 8)

# Read the metrics CSV file
csv_path = Path('/home/lum/AI_Zombie_Project/phase2/metrics.csv')
df = pd.read_csv(csv_path)

print("📊 Metrics Data Loaded:")
print(df.to_string())
print("\nData types:")
print(df.dtypes)
```

## Parse and Organize Data

Let's organize the metrics data by algorithm type and prepare it for analysis.

```python
# Organize data by scenario
print("📋 Data Organization:\n")

# Convert Winner column for better categorization
df['Winner_Type'] = df['Winner'].apply(lambda x: 'Prey' if x == 'prey' else 'Monster' if x == 'monster' else 'Timeout')
df['Algorithm_Type'] = df['Scenario'].apply(lambda x: 'MINIMAX (d=60)' if 'MINIMAX (d=60)' in x else 
                                                      'MINIMAX (d=3)' if 'MINIMAX (d=3)' in x else
                                                      'Ambush' if 'Ambush' in x else 'Evasive')

print("Summary by Winner Type:")
print(df['Winner_Type'].value_counts())
print("\nSummary by Algorithm Type:")
print(df['Algorithm_Type'].value_counts())

# Display organized data
print("\n✅ Data is ready for analysis!")
```

## Algorithm Performance Comparison

Compare key metrics across different algorithm implementations using bar charts.

```python
# Create 3-subplot comparison figure
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Plot 1: Steps to Completion
colors = ['green' if x == 'Prey' else 'red' for x in df['Winner_Type']]
axes[0].bar(df['Scenario'], df['Steps'], color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)
axes[0].set_ylabel('Steps to Completion', fontweight='bold', fontsize=11)
axes[0].set_title('Game Duration (Steps)', fontweight='bold', fontsize=12)
axes[0].set_xticklabels(df['Scenario'], rotation=45, ha='right')
axes[0].grid(axis='y', alpha=0.3)
for i, v in enumerate(df['Steps']):
    axes[0].text(i, v + 20, str(v), ha='center', fontweight='bold', fontsize=10)

# Plot 2: Computation Time
axes[1].bar(df['Scenario'], df['Computation_Time_s'], color='steelblue', alpha=0.7, edgecolor='black', linewidth=1.5)
axes[1].set_ylabel('Time (seconds)', fontweight='bold', fontsize=11)
axes[1].set_title('Computational Time', fontweight='bold', fontsize=12)
axes[1].set_xticklabels(df['Scenario'], rotation=45, ha='right')
axes[1].grid(axis='y', alpha=0.3)
for i, v in enumerate(df['Computation_Time_s']):
    axes[1].text(i, v + 0.002, f'{v:.4f}s', ha='center', fontweight='bold', fontsize=9)

# Plot 3: Total Nodes Evaluated
colors_nodes = ['lightgray' if x == 0 else 'orange' for x in df['Total_Nodes_Evaluated']]
axes[2].bar(df['Scenario'], df['Total_Nodes_Evaluated'], color=colors_nodes, alpha=0.7, edgecolor='black', linewidth=1.5)
axes[2].set_ylabel('Nodes Evaluated', fontweight='bold', fontsize=11)
axes[2].set_title('AI Complexity (Path Evaluations)', fontweight='bold', fontsize=12)
axes[2].set_xticklabels(df['Scenario'], rotation=45, ha='right')
axes[2].grid(axis='y', alpha=0.3)
for i, v in enumerate(df['Total_Nodes_Evaluated']):
    axes[2].text(i, v + 100, str(v), ha='center', fontweight='bold', fontsize=9)

plt.tight_layout()
plt.show()

print("✅ Performance comparison visualization complete!")
```

## Win Rate Visualization by Strategy

Analyze success rates and outcomes for each algorithm variant.

```python
# Win rate analysis
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Pie chart of outcomes
winner_counts = df['Winner_Type'].value_counts()
colors_pie = ['#2ecc71', '#e74c3c', '#95a5a6']  # Green (Prey), Red (Monster), Gray (Timeout)
axes[0].pie(winner_counts.values, labels=winner_counts.index, autopct='%1.1f%%', 
            colors=colors_pie, startangle=90, textprops={'fontsize': 11, 'fontweight': 'bold'})
axes[0].set_title('Overall Outcome Distribution', fontweight='bold', fontsize=12)

# Bar chart of winner by scenario
winner_by_scenario = df.set_index('Scenario')['Winner_Type']
scenario_colors = ['green' if x == 'Prey' else 'red' if x == 'Monster' else 'gray' for x in winner_by_scenario]
axes[1].barh(range(len(winner_by_scenario)), [1]*len(winner_by_scenario), 
             color=scenario_colors, edgecolor='black', linewidth=1.5)
axes[1].set_yticks(range(len(winner_by_scenario)))
axes[1].set_yticklabels(winner_by_scenario.index, fontsize=10)
axes[1].set_xlabel('Winner', fontweight='bold', fontsize=11)
axes[1].set_title('Winner by Scenario', fontweight='bold', fontsize=12)
axes[1].set_xlim(0, 1)
axes[1].set_xticks([])

# Add legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='green', label='Prey Wins'),
                   Patch(facecolor='red', label='Monster Wins'),
                   Patch(facecolor='gray', label='Timeout')]
axes[1].legend(handles=legend_elements, loc='lower right')

plt.tight_layout()
plt.show()

print("✅ Win rate visualization complete!")
```
